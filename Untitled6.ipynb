```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
content=pd.read_csv('Content.csv')
reaction=pd.read_csv('Reactions.csv')
reaction_type=pd.read_csv('ReactionTypes.csv')
```

In content file:


*  There are null values in URL
*   Unnamed: 0, URL don't provide anything meaningful to the analysis
*  The column 'Type' should be renamed as 'Content_Kind'
* In 'Category' some values start from uppercase while some lowercase.



```python
content.drop(columns=['Unnamed: 0','URL','User ID'],inplace=True)
```

```python
content.rename(columns={'Type':'Content_Kind'},inplace=True)
```

Reaction File

```python
reaction.head()
```

```python
reaction.info()
```

In Reaction file:


*   User ID, Type, Datetime have NULL values. We shall remove the NULL values
*   Unnamed: 0 is useless for current analysis. We shall remove that
*   Datetime column has Dtype object. We shall change it to 'Datetime' Dtype



```python
reaction.dropna(inplace=True)
```

```python
reaction.info()
```

```python
reaction.drop(columns=['Unnamed: 0','User ID'],inplace=True)
```

```python
from datetime import datetime
```

```python
reaction['Datetime']=reaction['Datetime'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
reaction['Datetime']
```

```python
reaction.info()
```

Reaction_type

```python
reaction_type
```

In reaction_type file:
*  'Unnamed: 0' is meaningless for current analysis. We shall remove that

```python
reaction_type.drop(columns='Unnamed: 0',inplace=True)
```

```python
reaction_type.info()
```

Data Merging:
*  Reaction is the target file.

* In order to gain useful insights, we should merge Reaction file with Content file. The common column between them is 'Content ID'. Merging is done on that column.

* After the above merging, the resultant file will later be merged with 'Reaction Type' file, using column 'Type'

```python
df1=reaction.merge(content,on='Content ID',how='left')
```

```python
df1.info()
```

```python
df=df1.merge(reaction_type,on='Type',how='left')
```

```python
df.info()
```

Distribution of Sentiment based on Content_Kind

```python
plt.figure(figsize=(15,6))
ax=sns.countplot(data=df,x='Content_Kind',hue='Sentiment',palette='Greens')
plt.ylabel('Count of Distinct Content_Kind')
ax.legend(title='Sentiment', bbox_to_anchor=(1, 1), loc='upper left', fontsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.set_xlabel('Content Kind', fontsize=15)
ax.set_ylabel('Count of Content Kinds', fontsize=12)

#ax.set_title('Distribution of Sentiments Across Distinct Content_Kinds')
fig = ax.get_figure()
fig.savefig("Distribution of Sentiments Across Distinct Content_Kinds.png")
```
